```python
import pandas as pd
import numpy as np
```

```python
health = pd.read_csv('health_modified.csv', encoding = 'latin1')
```

```python
health.head()
```

```python
data = pd.read_csv('dataset.csv', encoding = 'latin1')
```

```python
data.head()
```

```python
data.head()
```

```python
data['FIPS'].shape
```

```python
fips = pd.read_csv('fipscode_state.csv')
```

```python
fips.head()
```

```python
fips = fips.iloc[:, :2]
```

```python
fips.head()
```

```python
fips = fips.rename(columns = {'STATE' : 'FIPS', 'STUSAB' : 'STATE'})
```

```python
fips.head()
```

```python
array = []
```

```python
for each in data.loc[:,'FIPS']:
        each = int(each // 1000)
        array.append(fips.loc[fips['FIPS'] == each, 'STATE'].iloc[0])
```

```python
array
```

```python
array = pd.DataFrame(array)
```

```python
array.head()
```

```python
array = array.rename(columns = {0:'STATE'})
```

```python
array.head()
```

```python
data.head()
```

```python
dataset = data.join(array, how = 'outer')
```

```python
dataset.head()
```

```python
lst = dataset.columns.tolist()
```

```python
lst
```

```python
lst = ['FIPS',
       'STATE',
       'N02650',
       'A02650',
       'Population Estimate, 2013',
       'Population Estimate, 2016',
       'PCT_DIABETES_ADULTS13',
       'PCT_OBESE_ADULTS13']
```

```python
dataset = dataset[lst]
```

```python
dataset.head()
```

```python
county = []
for each in dataset.loc[:,'FIPS']:
    county.append(health.loc[health['FIPS'] == each, 'County'].iloc[0])
```

```python
county
```

```python
len(county)
```

```python
county = pd.DataFrame(county)
```

```python
county.head()
```

```python
county = county.rename(columns = {0:'County'})
```

```python
county.head()
```

```python
final_dataset = dataset.join(county, how = 'outer')
```

```python
final_dataset.head()
```

```python
lst2 = final_dataset.columns.tolist()
```

```python
lst2
```

```python
lst2 = ['FIPS',
 'STATE',
 'County',
 'N02650',
 'A02650',
 'Population Estimate, 2013',
 'Population Estimate, 2016',
 'PCT_DIABETES_ADULTS13',
 'PCT_OBESE_ADULTS13']
```

```python
final_dataset = final_dataset[lst2]
```

```python
final_dataset.head()
```

```python
final_dataset.to_csv('Final_dataset.csv')
```
