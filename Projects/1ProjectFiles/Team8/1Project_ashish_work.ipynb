```python
# Ashish Kumar
# Extracting data from Data 2 CHSI
#########################################################################
# imports

import numpy as np
import pandas as pd
from os import path
import sys
```

```python
#
comprehensive_health_df1 = pd.read_excel(
        '2015-County-Health-Rankings-Data.xls',
        sheet_name='Additional Measure Data',
        header=0,
        index_col=None,
        skiprows=1,
        usecols='A,AD,AG,AN,AR,AV,AX,BV'
    )
```

```python
comprehensive_health_df1.head(5)
```

```python
comprehensive_health_df2 = pd.read_excel(
        '2015-County-Health-Rankings-Data.xls',
        sheet_name='Ranked Measure Data',
        header=0,
        index_col=None,
        skiprows=1,
        usecols='A,O,T,AA,AF,AJ,AN,AP,AU,AX,BG,BK,BV,BZ,CD'
    )
```

```python
comprehensive_health_df2.head(5)
```

```python
comprehensive_health_df_withna = pd.merge(comprehensive_health_df1, comprehensive_health_df2, on='FIPS')
```

```python
comprehensive_health_df_withna.shape
```

```python
comprehensive_health_df = comprehensive_health_df_withna.dropna()
```
