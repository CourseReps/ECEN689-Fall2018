# Data visualisation
This notebook is for visualising some of the incomre-related data from CDC.gov
The data sets can be found here: https://www.cdc.gov/nchs/nhis/shs/tables.htm

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.offsetbox import AnchoredText
```

```python
########################################################################
# TWEAK MATPLOTLIB FOR IEEE STYLE
mpl.rcParams['axes.titlesize'] = 8
mpl.rcParams['axes.labelsize'] = 6
mpl.rcParams['legend.fontsize'] = 6
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = 'Times New Roman'
mpl.rcParams['xtick.labelsize'] = 5
mpl.rcParams['ytick.labelsize'] = 5
mpl.rcParams['figure.figsize'] = (3.5, 1.64)
# mpl.rcParams['lines.markersize'] = 2
# mpl.rcParams['lines.markeredgewidth'] = 0.5
mpl.rcParams['figure.dpi'] = 300

# Font manager for text boxes
FONT_PROPERTIES = {'family': 'serif', 'size': 5}
```

### function to plot trends

```python
def plot_trends(fname = '2013_SHS_Sadness', plt_title='Trend w.r.t income'):
    csv_dtypes = {'Income (USD)':str}
    #change path to datasets/cdc_data later
    df = pd.read_csv('../datasets/cdc_data/'+fname+'.csv', encoding='cp1252', dtype=csv_dtypes)
    col_idx = df.columns
    plot_col = col_idx.tolist()
    print(plot_col)
    plot_col.remove('Income (USD)')
    x_vals = range(0, len(df['Income (USD)']))
    fig = plt.figure(figsize=[7.5, 1.64])
    ax = df.plot(x='Income (USD)', y=plot_col, kind='line', 
            style='.-',
            fontsize=5, rot=1, 
            title = plt_title)
            #colormap='plasma')
    ax.set_xlabel('Income (USD)')
    ax.set_ylabel('Percentage population')
    plt.xticks(x_vals, df['Income (USD)'])
    plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
    plt.savefig('cdc_images/'+fname + '.png', bbox_inches='tight')
    plt.savefig('cdc_images/'+fname + '.eps', type='eps', bbox_inches='tight')
```

## Self assessment of health

```python
plot_trends(fname='2013_SHS_self_assessment', plt_title='Self report of health by income')
```

## Cancer data

```python
plot_trends(fname='2013_SHS_Cancer', plt_title='Cancer statistics by income')
```

## Mental health
This shows feelings of sadness, hoplessness etc. w.r.t income

```python
plot_trends(fname='2013_SHS_Sadness', plt_title='Feelings of sadness, hopelessness, worthlessness all or most of the time')
```

## General diseases

```python
plot_trends(fname='2013_SHS_general_diseases', plt_title='Common diseases by income')
```
