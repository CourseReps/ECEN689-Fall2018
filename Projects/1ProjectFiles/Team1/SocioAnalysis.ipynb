** Part1: Explore the trends of wealth and with socioeconomic groups**

```
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.patches as mpatches
```

```
x=state_social_df.MEDHHINC15	#A02650
y=state_social_df.PCT_DIABETES_ADULTS13
z=state_social_df.PCT_OBESE_ADULTS13

#area = state_social_df.PCT_OBESE_ADULTS08
fig, ax = plt.subplots()
im=ax.scatter(x, y,c=z,cmap=plt.cm.jet)
plt.xlabel('median household income in $', fontweight='bold')
plt.ylabel('% of Population with Diabetes 2013', fontweight='bold')
cbar=fig.colorbar(im, ax=ax)
cbar.set_label('% population with obesity 2013', fontweight='bold')


for i, txt in enumerate(state_social_df.index.values):
    ax.annotate(txt, (x[i], y[i]))
    
fig.savefig('image.png')
```

```
from scipy.stats import linregress
print(linregress(state_social_df.MEDHHINC15.values, state_social_df.PCT_DIABETES_ADULTS13.values))
print(linregress(state_social_df.MEDHHINC15.values, state_social_df.PCT_OBESE_ADULTS13.values))
```

**Seperate cluster analysis**

```
cluster_data =state_social_df.iloc[:,[9,19]].values
cluster_data
```

```
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(cluster_data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)

plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
plt.savefig('image4.png')
files.download('image4.png')
```

```
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(cluster_data)
```

```
plt.scatter(cluster_data[y_kmeans == 0, 0], cluster_data[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(cluster_data[y_kmeans == 1, 0], cluster_data[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(cluster_data[y_kmeans == 2, 0], cluster_data[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('cluster plot')
plt.xlabel('median household income in $')
plt.ylabel('% Diabetic population 2013')
plt.legend()

x=state_social_df.iloc[:,9].values
y=state_social_df.iloc[:,18].values 

for i, txt in enumerate(state_social_df.index.values):
    plt.annotate(txt, (x[i], y[i]))
    
plt.show()


```
