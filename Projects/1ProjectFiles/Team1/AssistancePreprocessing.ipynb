```
import io 
assist_df = pd.read_csv(io.StringIO(uploaded_assist['assistancesheet.csv'].decode('cp1252')))
healthirs_df = pd.read_csv(io.StringIO(uploaded_healthirs['Health_IRS.csv'].decode('cp1252')),index_col=0)
```

```
assist_df=assist_df.drop(['County'],axis=1)
assist_df.head(3)
```

```
healthirs_df=healthirs_df.drop(['County'],axis=1)
healthirs_df.head(3)
```
