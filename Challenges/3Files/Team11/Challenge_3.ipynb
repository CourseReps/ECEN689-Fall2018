```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error

path = r"D:/Study/3rd Sem/ECEN 689/Challenge 3"
os.chdir(path)

population_training_df = pd.read_csv('population_training_kaggle.csv', encoding='cp1252').dropna(axis=0)
population_testing_df = pd.read_csv('population_testing_kaggle.csv', encoding='cp1252').dropna(axis=0)
print(population_training_df.shape)
print(population_testing_df.shape)
population_training_df.drop(['Id'], axis=1, inplace=True)
population_testing_df.drop(['Id'], axis=1, inplace=True)

population_training_matrix = population_training_df.values
population_testing_matrix = population_testing_df.values
print(population_training_matrix.shape)
print(population_testing_matrix.shape)
```

```python
## Lists to save parameters
lst = [[] for i in range(population_training_matrix.shape[1])]
num_coef = [[] for i in range(population_training_matrix.shape[1])]
prediction = [[] for i in range(population_training_matrix.shape[1])]

## Lasso function definition
def lasso_reg (population_training_matrix,population_testing_matrix,alpha):
    for i in range(population_training_matrix.shape[1]):
        data_train = population_training_matrix
        data_test = population_testing_matrix
        while alpha > 0:
            alpha = round(alpha,2)
            lassoreg = Lasso(alpha=alpha,normalize=True, max_iter=1e8)
            lassoreg.fit(np.delete(data_train,i,1),data_train[:,i])
            y_predict = lassoreg.predict(np.delete(data_test,i,1))
            alpha = alpha*1.05
           
            
            if np.count_nonzero(lassoreg.coef_)<=5:
                num_coef[i].append([np.count_nonzero(lassoreg.coef_)])
                lst[i].extend([lassoreg.coef_])
                prediction[i].extend([y_predict])
                print(alpha)
                print(i)
                print(np.count_nonzero(lassoreg.coef_))
                alpha = 1
                break
    return lst,num_coef,prediction
    
        
```

```python
## Running Lasso function
alpha = 2
(coefficients,n_coef,prediction) = lasso_reg(population_training_matrix,population_testing_matrix,alpha)
```

```python
## Create dataframe of predictions
col_names = population_testing_df.columns
prediction_list = [item for sublist in prediction for item in sublist]
prediction_df = pd.DataFrame(prediction_list)
prediction_df = prediction_df.T
prediction_df.columns = col_names
year_pred = np.linspace(np.int(2000),np.int(2016),17)

prediction_df.insert(loc = 0,column = 'Id',value = year_pred)
prediction_df.Id = prediction_df.Id.astype(int)
```

```python
## Change path and save the predictions
prediction_df.Id = prediction_df.Id.astype(int)
path = r"D:/Study/3rd Sem/ECEN 689/Challenge 3"
os.chdir(path)
prediction_df.to_csv(path+"prediction2.csv")
```

```python
## Change path and save the coefficients
coefficients_list = [item for sublist in coefficients for item in sublist]
for i in range(len(coefficients_list)):
    coefficients_list[i] = np.insert(coefficients_list[i],i,0)
```

```python
## Change path and save coefficients
new_coeff_df = pd.DataFrame(coefficients_list,columns = col_names,index = col_names)
new_coeff_df = new_coeff_df.T
new_coeff_df.to_csv(r"D:/Study/3rd Sem/ECEN 689/Challenge 3"+"coefficients.csv")
```
