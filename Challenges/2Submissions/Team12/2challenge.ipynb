```
import pandas as pd
from google.colab import files 
import numpy as np

```

**Load the training data**

```
uploaded_train = files.upload()
```

**Load the test data**

```
uploaded_test = files.upload()
```

```
for fn1 in uploaded_train.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn1, length=len(uploaded_train[fn1])))
  
for fn2 in uploaded_test.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn2, length=len(uploaded_test[fn2])))
```

```
import io 
train_data= pd.read_csv(io.StringIO(uploaded_train['mnist_train.csv'].decode('utf-8')))
train_data.head()
```

```
test_data= pd.read_csv(io.StringIO(uploaded_test['mnist_test.csv'].decode('utf-8')))
test_data.head()
```

**Preprocessing the test and training data**

```
x_train=train_data.drop(['Id','Category'], axis=1)
y_train=train_data.Category

x_test=test_data.drop(['Id'], axis=1)
```

**Building the classifiers**

```
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

**Train KNN with 5 neighbors**

```
CLF_KNN = KNeighborsClassifier(n_neighbors=5)
```

```
CLF_KNN.fit(x_train, y_train)
```

**Applying scaling and PCA for Logistic Regression**

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train_Scaled = sc.fit_transform(x_train)
X_test_Scaled = sc.transform(x_test)
```

**Using 300 pca components which explains 93% of the variance of the data.**

```
from sklearn.decomposition import PCA
n_components=300
pca=PCA(n_components=n_components,svd_solver='randomized',whiten=True).fit(X_train_Scaled)
X_train_pca=pca.transform(X_train_Scaled)
X_test_pca=pca.transform(X_test_Scaled)
explained_var=pca.explained_variance_ratio_
print(np.sum(explained_var))

```

**Train the logistic regression classifier**

```
CLF_Logit = LogisticRegression(solver = 'lbfgs',penalty='l2',C=0.01,random_state=0)
CLF_Logit.fit(X_train_pca, y_train)
```

**Get the training accuracy and predict logistic regression test labels**

```
from sklearn import metrics
predict_logit_train=CLF_Logit.predict(X_train_pca)
score_logit=metrics.accuracy_score(y_train,predict_logit_train)
print(score_logit)
predict_logit_test=CLF_Logit.predict(X_test_pca)
coef_logit=CLF_Logit.coef_
```

**Get the knn predicted labels**

```
predict_test_knn=CLF_KNN.predict(x_test)
```
