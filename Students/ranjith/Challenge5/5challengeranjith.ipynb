````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedShuffleSplit

student_identity = 'ranjith'

filenameToRead = '5challenge_training_' + student_identity + '.csv'
training_df = pd.read_csv(filenameToRead)
test_file = '5challenge_testing_' + student_identity + '.csv'
testing_df = pd.read_csv(test_file)
print(training_df.head())

X1 = training_df[['Feature 0', 'Feature 1']].values
Y1 = training_df[['Class']].values
Y1
````

````python

Y1=Y1.ravel()
Y1
````

````python
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')

plt.show()
````

````python
model_linear = svm.SVC(kernel='linear') 

model_linear.fit(X1, Y1)
model_linear.score(X1, Y1)
````

````python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X=X1,y=Y1,clf=model_linear,legend=2)
plt.xlabel('Feature 0',size=20)
plt.ylabel('Feature 1',size=20)
plt.title('Linear SVM Classification',size=30)
````

````python
model_poly = svm.SVC(kernel='poly')  
model_poly.fit(X1, Y1)
model_poly.score(X1, Y1)
````

````python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X=X1,y=Y1,clf=model_poly,legend=2)
plt.xlabel('Feature 0',size=20)
plt.ylabel('Feature 1',size=20)
plt.title('Poly SVM Classification',size=30)
````

````python
model_sigmoid = svm.SVC(kernel='sigmoid')  
model_sigmoid.fit(X1, Y1)  
model_sigmoid.score(X1, Y1)
````

````python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X=X1,y=Y1,clf=model_sigmoid,legend=2)
plt.xlabel('Feature 0',size=20)
plt.ylabel('Feature 1',size=20)
plt.title('Sigmoid SVM Classification',size=30)
````

````python
model_rbf.fit(X1, Y1)
model_rbf.score(X1, Y1)
````

````python

from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X=X1,y=Y1,clf=model_rbf,legend=2)
plt.xlabel('Feature 0',size=20)
plt.ylabel('Feature 1',size=20)
plt.title('RBF SVM Classification',size=30)
````

````python
range_of_c_values = np.logspace(-2, 10, 13)
range_of_gamma_values = np.logspace(-9, 3, 13)
grid_of_parameters = dict(gamma=range_of_gamma_values, C=range_of_c_values)
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
OptGrid = GridSearchCV(svm.SVC(), param_grid=grid_of_parameters, cv=None)
OptGrid.fit(X1, Y1)

print("The best parameters are %s with a score of %0.2f"
      % (OptGrid.best_params_, OptGrid.best_score_))
````

````python
#The best parameters are {'C' : 1000000.0, 'gamma' : 0.1}
model_rbf1 = svm.SVC(kernel='rbf',gamma=0.1,C=1000000)
````

````python
model_rbf1.fit(X1, Y1)
model_rbf1.score(X1, Y1)
````

````python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X=X1,y=Y1,clf=model_rbf1,legend=2)
plt.xlabel('Feature 0',size=20)
plt.ylabel('Feature 1',size=20)
plt.title('RBF SVM Classification',size=30)
````
