### Metric
 * Average $L_2$ norm between the true and estimated parameters.

````python
import pandas as pd
import numpy as np

student_identity = 'ranjith'
filename = '1challenge1activity_' + student_identity + '.csv'
my_data_frame = pd.read_csv(filename)
my_data_frame.head(11)
````

````python
cropped_data_frame=my_data_frame[my_data_frame.columns[3:14]]
my_data_frame['Mean'] = cropped_data_frame.mean(axis=1)
my_data_frame['Variance'] = cropped_data_frame.var(axis=1)
my_data_frame.head(11)
````

````python
my_data_frame.to_csv('1challenge1activity_answer' + student_identity + '.csv')
````
