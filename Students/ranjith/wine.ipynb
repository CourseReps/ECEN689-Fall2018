```
model.add(tf.keras.layers.Dense(1,activation='linear'))
```

```
adam1=keras.optimizers.Adadelta(lr=1.0, rho=0.95,epsilon=None, decay=0.0)
model.compile(optimizer=adam1,
              loss='mse',
              metrics=['mae'])
```
