### Metric

````python
import pandas as pd
import numpy as np

student_identity = 'riladdha'
filenameToRead = '1challenge1activity_' + 'riladdha' + '.csv'
filename = '1challenge1activity_' + student_identity + '.csv'
sample_df = pd.read_csv('1Files/' + filenameToRead)

sample_df['Mean']=sample_df.loc[:,'Sample 0': 'Sample 11'].mean(axis=1)
sample_df['Variance']=sample_df.loc[:,'Sample 0': 'Sample 11'].var(axis=1) 
sample_df.to_csv(filename, sep=',')

sample_df.head(5)
````
