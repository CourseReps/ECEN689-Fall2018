```
import pandas as pd
import numpy as np
import io
from keras.layers import Dense
from keras.models import Sequential
from keras.utils import np_utils
```

```
filetoRead='winequality-white-training.csv'
winetrain_df = pd.read_csv(io.StringIO(file_upload['winequality-white-training.csv'].decode('utf-8')))
winetrain_df=winetrain_df.dropna()
print(winetrain_df.shape)
winetrain_df
```

```
train_df=winetrain_df.drop(['Id','quality'],axis=1)
xtrain=train_df
xtrain.shape
xtrain
```

```
ytrain=pd.DataFrame()
ytrain=winetrain_df['quality']
ytrain.shape
ytrain
```

```
winetest_df = pd.read_csv(io.StringIO(file_upload['winequality-white-testing.csv'].decode('utf-8')))
xtest = winetest_df.drop(['Id'],axis=1)
xtest.shape
```

```
xtraining = np.array(xtrain.values.tolist(),dtype=float)
ytraining = np.array(ytrain.values.tolist(),dtype=float)
#ytraining
#xtraining

a= pd.DataFrame(np.zeros((1000, 1)),dtype=int)
ytesting = a
xtesting = np.array(xtest.values.tolist(),dtype=float)
ytesting = np.array(ytest.values.tolist(),dtype=float)
ytesting
```

```
import tensorflow as tf
import tensorflow.keras as keras
nadam=keras.optimizers.Nadam(lr=0.001, beta_1=0.9, beta_2=0.999,epsilon=1e-08,schedule_decay=0.004)
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(9, kernel_initializer="normal", activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(11, input_dim=11, kernel_initializer="normal", activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(1,activation='linear'))

model.compile(loss='mse', optimizer='nadam', metrics=['mse'])
```

```
ytest.shape
```

```

model.fit(xtraining,ytraining,epochs=500, batch_size=12)
```

```
ytesting=model.predict(xtesting)

```
