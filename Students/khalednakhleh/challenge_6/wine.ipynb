```
white_train = pd.read_csv("drive/My Drive/data_files/winequality-white-training.csv")
white_test = pd.read_csv("drive/My Drive/data_files/winequality-white-testing.csv")

print("White wine training data dimensions: " + str(white_train.shape))
print("White wine testing data dimensions: " + str(white_test.shape))

x_train = white_train.iloc[:, 1: 12].values
y_train = white_train.iloc[:, 12].values

y_train = keras.utils.to_categorical(y_train, dtype = "float32")

x_test = white_test.iloc[:, 1: 12].values
```

```
model = Sequential()
model.add(Dense(200, input_dim = 11, activation = "relu"))
model.add(Dense(100, kernel_initializer ="uniform", activation = "relu"))
model.add(Dense(50, kernel_initializer = "uniform", activation = "relu"))
model.add(Dense(25, kernel_initializer = "uniform", activation = "relu"))
model.add(Dense(10, kernel_initializer = "uniform", activation = "softmax"))

model.compile(loss = "mean_squared_error",
              optimizer = "adam",
              metrics=["accuracy"])

model.fit(x_train, y_train, epochs = 4000, batch_size = 10, verbose = 2)

model.save("wine_result.h5")
```

```
score = model.evaluate(x_train, y_train)
predictions = model.predict_classes(x_test)

print("\nModel accuracy: " + str(100 * round(score[1], 3)) + "%\n")

result = pd.DataFrame(data = predictions, columns = ["quality"])

result.index.name = "Id"

result.to_csv("drive/My Drive/data_files/wine_result.csv")

files.download("drive/My Drive/data_files/wine_result.csv" ) 

print(result)
```
