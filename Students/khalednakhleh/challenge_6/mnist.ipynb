```
train = pd.read_csv("drive/My Drive/data_files/mnist_train.csv", index_col = 0)
test = pd.read_csv("drive/My Drive/data_files/mnist_test.csv", index_col = 0)

print("Training data dimensions: " + str(train.shape))
print("Testing data dimensions: " + str(test.shape))

x_train = train.iloc[:, 1:784].values
y_train = train.iloc[:, 0].values

x_test = test.iloc[:, 0:783].values

y_train = keras.utils.to_categorical(y_train, dtype = "float32")
```

```
model = Sequential()
model.add(Dense(200, input_dim = 783, activation = "relu"))
model.add(Dense(175, kernel_initializer = "uniform", activation = "relu"))
model.add(Dense(150, kernel_initializer = "uniform", activation = "relu"))
model.add(Dense(125, kernel_initializer = "uniform", activation = "relu"))
model.add(Dense(100, kernel_initializer = "uniform", activation = "relu"))
model.add(Dense(75, kernel_initializer = "uniform", activation = "relu"))
model.add(Dense(50, kernel_initializer = "uniform", activation = "relu"))
model.add(Dense(25, kernel_initializer = "uniform", activation = "relu"))
model.add(Dense(10, kernel_initializer = "uniform", activation = "softmax"))

model.compile(loss = "mean_squared_error",
              optimizer = "adam",
              metrics = ["accuracy"])

model.fit(x_train, y_train, epochs = 50, batch_size = 100, verbose = 2)

model.save("mnist_result.h5")
```

```
score = model.evaluate(x_train, y_train)
predictions = model.predict_classes(x_test)

print("\nModel accuracy: " + str(100 * round(score[1], 3)) + "%\n")

result = pd.DataFrame(data = predictions, columns = ["Category"])

result.index += 1
result.index.name = "Id"

result.to_csv("drive/My Drive/data_files/mnist_result.csv")
files.download("drive/My Drive/data_files/mnist_result.csv" ) 

print(result)
```
