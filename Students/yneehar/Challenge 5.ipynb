```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import os
path = 'C:\\Users\\Neehar\\Desktop\\689 Project\\challenge 5'
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.metrics import classification_report, confusion_matrix
from mlxtend.plotting import plot_decision_regions
```

```python
training_df = pd.read_csv(path + '\\5challenge_training_yneehar.csv')
testing_df = pd.read_csv(path + '\\5challenge_testing_yneeharold.csv')
```

```python
print(training_df.head())

X1 = training_df[['Feature 0', 'Feature 1']].values
Y1 = training_df['Class'].values
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')
plt.xlabel('X0', size=14)
plt.ylabel('X1', size=14)
plt.show()
```

```python
plt.clf()
plt.hist(X1[:,0], bins= 25, alpha = 0.7)
plt.hist(X1[:,1], bins= 25, alpha = 0.6)
plt.legend(['Feature 0', 'Feature 1'])
plt.show()
```

```python
svclassifier = svm.SVC(kernel='rbf', gamma = 1, C=0.15)
svclassifier.fit(X1, Y1)
y_pred = svclassifier.predict(X1)
print(confusion_matrix(Y1, y_pred))  
print(classification_report(Y1, y_pred))
```

```python
plot_decision_regions(X=X1, 
                      y=Y1,
                      clf=svclassifier, 
                      legend=1)
plt.xlabel('X0', size=14)
plt.ylabel('X1', size=14)
plt.title('SVM Decision Region Boundary', size=16)
plt.show()
```
