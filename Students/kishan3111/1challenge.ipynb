````python
import pandas as pd
import numpy as np

student_identity = 'kishan3111'
filename = '1challenge1activity_' + student_identity + '.csv'
sample_df = pd.read_csv('D:/Study/3rd Sem/ECEN 689/ECEN689-Fall2018/Challenges/1Files/' + filename)
````

````python
sample_df
````

````python
# Remove 1st Column
sample_df = sample_df.iloc[:, 1:]
````

````python
sample_df
````

````python
# Evaluating mean for each row and assign all the values of mean to Mean column

sample_df['Mean'] = sample_df.iloc[:, 2:].mean(axis=1)
````

````python
# Evaluating variance for each row and assign all the values of variance to Variance column

sample_df['Variance'] = sample_df.iloc[:, 2:].var(axis=1)
````
