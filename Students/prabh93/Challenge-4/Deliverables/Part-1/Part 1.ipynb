```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
wine_training_df1 = pd.read_csv('C:/Users/prabh/Desktop/Challenge 4/Part-1/winequality-white-training.csv')
wine_testing_df1 = pd.read_csv('C:/Users/prabh/Desktop/Challenge 4/Part-1/winequality-white-testing.csv')
```

```python
wine_training_df = wine_training_df1.drop(labels='Id', axis=1) 
```

```python
wine_training_df.head(5)
```

```python
wine_testing_df = wine_testing_df1.drop(labels='Id', axis=1) 
```

```python
wine_testing_df.head(5)
```

```python
x_train = wine_training_df.drop('quality', axis=1)
```

```python
x_train.head(5)
```

```python
lm = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=1, normalize=True)
lm
```

```python
lm.fit(x_train,wine_training_df.quality)
```

```python
x_test = wine_testing_df
```

```python
y_pred = lm.predict(x_test)
```

```python
y_pred_df = pd.DataFrame(y_pred,columns=["quality"])
```

```python
y_pred_df.head(10)
```

```python
y_pred_df.insert(loc = 0,column = 'Id',value = wine_training_df1.Id)
```

```python
y_pred_df.head(10)
```

```python
y_pred_df.to_csv("D:/Study/3rd Sem/ECEN 689/Challenge 4 - Wine Testing/Part 1/" + "winequality-white-solution.csv",index=False)
```
