```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
white_training_df1 = pd.read_csv('D:/Study/3rd Sem/ECEN 689/Challenge 4 - Wine Testing/Part 1/winequality-white-training.csv')
white_testing_df1 = pd.read_csv('D:/Study/3rd Sem/ECEN 689/Challenge 4 - Wine Testing/Part 1/winequality-white-testing.csv')
```

```python
white_training_df = white_training_df1.drop(labels='Id', axis=1) 
white_testing_df = white_testing_df1.drop(labels='Id', axis=1)
```

```python
red_training_df1 = pd.read_csv('D:/Study/3rd Sem/ECEN 689/Challenge 4 - Wine Testing/Part 3/winequality-red-training.csv')
red_testing_df1 = pd.read_csv('D:/Study/3rd Sem/ECEN 689/Challenge 4 - Wine Testing/Part 3/winequality-red-testing.csv')
```

```python
red_training_df = red_training_df1.drop(labels='Id', axis=1) 
red_testing_df = red_testing_df1.drop(labels='Id', axis=1)
```

```python
x_train = white_training_df.drop('quality', axis=1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm
```

```python
lm.fit(x_train,white_training_df.quality)
```

```python
x_test = red_testing_df
```

```python
y_pred = lm.predict(x_test)
```

```python
y_pred_df = pd.DataFrame(y_pred,columns=["quality"])
```

```python
y_pred_df.insert(loc = 0,column = 'Id',value = red_training_df1.Id)
```
