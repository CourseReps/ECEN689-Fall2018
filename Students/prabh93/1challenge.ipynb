````python
import pandas as pd
import numpy as np

student_identity = 'prabhtamu'
filenameToRead = '1challenge1activity_' + 'prabhtamu' + '.csv'
filename = '1challenge1activity_' + student_identity + '.csv'
sample_df = pd.read_csv('D:/Semester-3/ECEN 689/ECEN689-Fall2018/Challenges/1Files/' + filenameToRead)
````

````python
sample_df.head() #first five rows by default
````

````python
sample_df=sample_df.iloc[:,1:] #slicing all rows and omitting first column
````

````python
sample_df.head()
````

````python
sample_df["Mean"]=sample_df.iloc[:,2:].mean(axis=1) #mean across rows starting from 2nd column
````

````python
sample_df.head()
````

````python
sample_df["Variance"]=sample_df.iloc[:,2:].var(axis=1) #variance across rows starting from 2nd column
````
