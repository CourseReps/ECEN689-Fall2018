```python
import pandas as pd
import numpy as np

from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error

population_training_df = pd.read_csv('D:\\Semester-3\\ECEN 689\\Challenge 3\\population_training.csv', encoding='cp1252').dropna(axis=0)
population_testing_df = pd.read_csv('D:\\Semester-3\\ECEN 689\\Challenge 3\\population_testing.csv', encoding='cp1252').dropna(axis=0)
population_testing_kaggle_df = pd.read_csv('D:\\Semester-3\\ECEN 689\\Challenge 3\\population_testing_kaggle.csv', encoding='cp1252').dropna(axis=0)
```

```python
names = list(population_training_df['Country Name'].values)
Id=population_testing_kaggle_df['Id']
```

```python
df = population_testing_df
```

```python
df_test = df.loc[df['Country Name'] != 'Kuwait']
```

```python
df_train = population_training_df
```

```python
df_train.drop(['Country Name'], axis=1, inplace=True)
df_test.drop(['Country Name'], axis=1, inplace=True)

population_training_matrix = df_train.values
population_testing_matrix = df_test.values
print(population_training_matrix.shape)
print(population_testing_matrix.shape)
```

```python
print(df_train.head(5))
```

```python
population_matrix = population_training_matrix[1:]
population_vector = population_training_matrix[0]
population_t = population_testing_matrix[1:]
population_t_vec = population_testing_matrix[0]
```

```python
alphas = 10**np.linspace(10,-2,1000)*0.5
```

```python
#Applying Lasso Regression and looping over multiple alphas
ridge = Lasso(normalize = True)
error = []
pred_final = []
coef_final = []
coef = []

for i in range(0,258):
    #print(i)
    population_vec = list(df_train[df_train.index == i].values.transpose())
    population_matrix = df_train[df_train.index != i].values
    population_vec_test = list(df_test[df_test.index == i].values.transpose())
    population_matrix_test = df_test[df_test.index != i].values
    #coef=[]
    minimum = float('inf')
    for a in alphas:
        ridge.set_params(alpha = a)
        ridge.fit(population_matrix.transpose(), population_vec)
        prediction = ridge.predict(population_matrix_test.transpose())
        error = mean_squared_error(prediction,population_vec_test)
        count = np.count_nonzero(ridge.coef_)
        if count <= 5 and count >2:
            if error < minimum:
                coef = ridge.coef_
                minimum = error
                prediction = ridge.predict(population_matrix_test.transpose())
    
    coef1 = []
    for k in range(i):
        coef1.append(coef[k])
    coef1.append(0)
    for k in range(i, 257):
        coef1.append(coef[k])
    coef_final.append(coef1) 
    pred_final.append(prediction)
```

```python
coef_arr = np.array(coef_final)
df1 = pd.DataFrame(data = coef_arr, columns = names, index = names)
```

```python
df1 = df1.transpose()
print(df1.shape)
print(df1.head())
```

```python
df1.to_csv('population_parameters.csv')
```

```python
#Converting to a dataframe
pred_arr = np.array(pred_final)

df = pd.DataFrame(data = pred_arr,columns = Id ,index=names)
df = df.transpose()
print(df.shape)
```
