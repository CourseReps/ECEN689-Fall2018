````python
import pandas as pd
import numpy as np

student_identity = 'alitamu3'
filename = '1challenge1activity_' + student_identity + '.csv'
sample_df = pd.read_csv('C:/Users/Ali/Documents/GitHub/ECEN689-Fall2018/Challenges/1Files/'+ filename)
sample_df.head(3)
````

````python
# Calculating the mean and Variance for each row
sample_df.loc[:,'Mean'] = sample_df.loc[:,'Sample 0':].mean(axis=1)
sample_df.loc[:,'Variance'] = sample_df.loc[:,'Sample 0':].var(axis=1)
````

````python
#Writing to respective csv file
sample_df.loc[:,'Mean':].to_csv(path_or_buf= 'C:/Users/Ali/Documents/GitHub/ECEN689-Fall2018/Challenges/1Files/'+ filename)
````
