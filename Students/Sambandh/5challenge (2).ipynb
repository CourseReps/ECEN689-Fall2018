````python
from sklearn import svm
# figure number
fignum = 1
#from sklearn.cross_validation import KFold, cross_val_score
#k_fold = KFold(len(Y1), n_folds=10, shuffle=True, random_state=0)
# fit the model
for kernel in ('linear', 'poly', 'rbf'):
    clf = svm.SVC(kernel=kernel, C=3, gamma=0.5)
    clf.fit(X1, Y1)
    #print (cross_val_score(clf, X1, Y1, cv=k_fold, n_jobs=1))
    print (clf.score(X1,Y1))
````

Parameters used in my SVM code:

C: It is the regularization parameter, C, of the error term.

Kernel: It specifies the kernel type to be used in the algorithm. It can be ‘linear’, ‘poly’ and ‘rbf’

Degree: It is the degree of the polynomial kernel function (‘poly’) and is ignored by all other kernels. The default value is 3.

Gamma: It is the kernel coefficient for ‘rbf’, ‘poly’, and ‘sigmoid’. If gamma is ‘auto’, then 1/n_features will be used instead.

I tried using different 'gamma' values ranging from 0.5 to 10 with some higher values of 50 and 100 too.

Then, The best result with the training data went on to occur at value of gamma = 100 ( Radial Kernel Function) which was 98% classification accuracy on the training data but
that seemed to overfit the model.
Therefore, I went on to use the 'auto' value for gamma and degree of polynomial = 3.

I went on to keep the same set of parameters while experimenting over the testing dataset and eventually, the classes were recorded over the testing dataset.

The Radial basis function kernel, also called the RBF kernel, or Gaussian kernel,
is a kernel that is in the form of a radial basis function (more specifically, a Gaussian
function). The RBF kernel is defined as
KRBF(x, x0) = exp (−γ ||x − x'||^2)
where γ is a parameter that sets the “spread” of the kernel.

````python
# plot the line, the points, and the nearest vectors to the plane
for kernel in ('linear', 'poly', 'rbf'):
    
    clf = svm.SVC(kernel=kernel, C=3, gamma=0.5)
    clf.fit(X1, Y1)
    plt.figure(fignum, figsize=(4, 3))
    plt.clf()

    plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=80, facecolors='none', zorder=10, edgecolors='k')
    #plt.scatter(X1[:, 0], X1[:, 1], s=80, facecolors='none', zorder=10, cmap=plt.cm.Paired, edgecolors='k')
    plt.axis('tight')
    x_min = -3
    x_max = 3
    y_min = -3
    y_max = 3

    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
    Z = clf.decision_function(np.c_[XX.ravel(), YY.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(XX.shape)
    plt.figure(fignum, figsize=(4, 3))
    plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
    plt.contour(XX, YY, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'],
                levels=[-.5, 0, .5])

    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)

    plt.xticks(())
    plt.yticks(())
    fignum = fignum + 1
plt.show()
````
