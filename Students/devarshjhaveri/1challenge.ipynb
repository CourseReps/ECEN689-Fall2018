````python
import pandas as pd
import numpy as np

student_identity = 'devarshjhaveri'
filenameToRead = '1challenge1activity_' + student_identity + '.csv'
filename = '1challenge1activity_' + student_identity + '.csv'
sample_df = pd.read_csv('/Users/devarshjhaveri/Desktop/Courses/ECEN689-Fall2018/Challenges/1Files/' + filenameToRead)

sample_df.head(5)
````

````python
sample_df['Mean'] = np.mean(sample_df[sample_df.columns[3:15]], axis=1)
sample_df['Variance'] = np.var(sample_df[sample_df.columns[3:15]],ddof=1, axis=1)
sample_df.head(5)
````
