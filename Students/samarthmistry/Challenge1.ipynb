```python
import numpy as np
import pandas as pd

student_identity = 'samarth'
filename = '1challenge1activity_' + student_identity + '.csv'

df = pd.read_csv('1challenge1activity_samarth.csv', index_col=0)
df.head()
```

```python
df['Mean'] = np.mean(df[df.columns[2:13]], axis=1)
df.head()
```

```python
df['Variance'] = np.var(df[df.columns[2:13]], axis=1, ddof=1)
df.head()
```

```python
df.to_csv(filename)
```
